# Extraer CSV

```python
import pandas as pd
import logging

def extract_data(filepath):
    try:
        df = pd.read_csv(filepath)
        logging.info("Datos cargados correctamente")
        return df

    except FileNotFoundError:
        logging.error("Archivo no encontrado")
        raise

    except Exception as e:
        logging.error(f"Error inesperado: {e}")
        raise
```

# Limpieza

```python
#revision de estructura
df.info()
```

## Renombrar columnas

```python
#Renombramos columnas
#Calories from\nFat
#Total Fat\n(g)
#Saturated Fat\n(g)

df.columns = (
    df.columns
    .str.replace("\n", "_")
    .str.replace(" ", "_")
    .str.replace("(", "")
    .str.replace(")", "")
)
```

```python
df.info()
```

## Limpiar duplicados

```python
df.drop_duplicates(inplace=True)
```

## Manejo de nulos

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

# Cambiar tipo de las variables

```python
columnas_numericas = [
    'Calories', 'Sodium__mg', 'Total_Fat_g', 'Calories_from_Fat', 
    'Saturated_Fat_g', 'Trans_Fat_g', 'Cholesterol_mg', 'Carbs_g', 
    'Fiber_g', 'Sugars_g', 'Protein_g', 'Weight_Watchers_Pnts'
]

for col in columnas_numericas:
    df[col] = pd.to_numeric(
        df[col].astype(str).str.replace('\xa0', '', regex=False).str.replace('<', '', regex=False), 
        errors='coerce')
```

## Calorias negativas

```python
df = df[df["Calories"] >= 0]
```

## Sodio negativo

```python
df = df[df["Sodium__mg"] >= 0]
```

## Grasas negativas

```python
df = df[df["Total_Fat_g"] >= 0]
```

## Crear variables nuevas

```python
def calorie_category(x):

    if x < 300:
        return "Baja"

    elif x < 600:
        return "Media"

    else:
        return "Alta"

df["Calorie_Category"] = df["Calories"].apply(calorie_category)
```
