# Extraer CSV

```python
import pandas as pd
import logging

def extract_data(filepath):
    try:
        df = pd.read_csv(filepath)
        logging.info("Datos cargados correctamente")
        return df

    except FileNotFoundError:
        logging.error("Archivo no encontrado")
        raise

    except Exception as e:
        logging.error(f"Error inesperado: {e}")
        raise
```

# Limpieza

```python
#revision de estructura
df.info()
```

## Renombrar columnas

```python
#Renombramos columnas
#Calories from\nFat
#Total Fat\n(g)
#Saturated Fat\n(g)

df.columns = (
    df.columns
    .str.replace("\n", "_")
    .str.replace(" ", "_")
    .str.replace("(", "")
    .str.replace(")", "")
)
```

## Limpiar duplicados

```python
df.drop_duplicates(inplace=True)
```

## Manejo de nulos

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

## Calorias negativas

```python
df = df[df["Calories"] >= 0]
```

## Sodio negativo

```python
df = df[df["Sodium_mg"] >= 0]
```

## Grasas negativas

```python
df = df[df["Total_Fat_g"] >= 0]
```

## Crear variables nuevas

```python
def calorie_category(x):

    if x < 300:
        return "Baja"

    elif x < 600:
        return "Media"

    else:
        return "Alta"

df["Calorie_Category"] = df["Calories"].apply(calorie_category)
```
